# 03. Exploring Relationships

## Correlations

```python
# Create a correlation matrix
df_football.corr()
```

```python
# Creating subset of dataframe
sub = df_football[['Goals', 'Non-Penalty Goals', 'Penalty Goals', 'Penalty Attempts', 'Shots', 'Shots on Target', 'xG', 'npxG']]
```

```python
sub
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(20,20))

# Create the correlation heatmap in seaborn
corr = sns.heatmap(sub.corr(), annot = True, ax = ax) 
```

I was hoping to find a strong correlation between goals and shots on target, but at 0.2, this is quite a weak correlation. This indicates although a shot may be on target, the strike is not of high enough quality to beat the goalkeeper. Naturally, there is a strong correlation between 'expected goals' (xG) and actual goals (Goals). There is also a very strong correlation between penalties attempted and penalties scored, showing a high probability of scoring a penalty if you take one.

## Scatterplots

```python
# Creating scatterplot goals vs non-penalty goals
sns.lmplot(x = 'Goals', y = 'Non-Penalty Goals', data = df_football)
```

Understandably there is a strong correlation between these to variables. Goals will always be either equal to or higher than non-penalty goals, and this scatterplot shows that as goals increase, the amount of non-penalty goal increase too. This indicates that on the whole (minus a few outliers), penalties are only a small percentage of a players overall goals.

## Categorical Plot

```python
# Creating histogram to visualise 'goals' distribution
sns.histplot(df_football['Goals'], bins = 10, kde = True)
```

```python
# Create low goalscorer category
df_football.loc[df_football['Goals'] < 5, 'Goalscorer category'] = 'Low goalscorer'
```

```python
# Create medium goalscorer category
df_football.loc[(df_football['Goals'] >= 5) & (df_football['Goals'] < 15), 'Goalscorer category'] = 'Medium goalscorer'
```

```python
# Create high goalscorer category
df_football.loc[df_football['Goals'] >= 15, 'Goalscorer category'] = 'High goalscorer'
```

```python
# Creating categorical plot - Goals vs xG
sns.set(style="ticks")
sns.scatterplot(x = 'xG', y='Goals', hue="Goalscorer category", data=df_football)
```

The higher the xG, the higher the amount of goals scored.
